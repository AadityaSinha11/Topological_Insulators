```python
import matplotlib.pyplot as plt
import numpy as np


#distance between lattice sites
a =1

#Number of lattice sites
N= 100

# total length of the chain
L=N*a

#onsite energy
e = 0

#hopping potential
t =0.5

k=np.linspace(-np.pi/a, np.pi/a,90)
exact = e -2.*t*np.cos(k*a)

xticks = np.linspace(-np.pi/a, np.pi/a,9)
xlabels = ['' for k in xticks]
xlabels[0] = '$-\\frac{\pi}{a}$'
xlabels[-1] = '$\\frac{\pi}{a}$'
fig, axs = plt.subplots()
axs.set_xlim(-np.pi/a,np.pi/a)
axs.set_title('Tight Binding Model')
axs.set_ylabel('$E$')
axs.set_xlabel('$ka$')
axs.set_yticks([e-2.*t, e-t, e, e+t, e+2.*t])
axs.set_yticklabels(['$E_0-2t$', '$E_0-t$', '$E_0$', '$E_0+t$', '$E_0+2t$'])
axs.set_xticks(xticks)
axs.set_xticklabels(xlabels)
axs.grid(True)
axs.plot(k, exact)
axs.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define constants
k = 2 * np.pi / 5  # Wave vector
L = 10  # Periodic lattice length

# Define a periodic function u_k(x)
def u_k(x):
    return np.exp(1j * k * x)  # plane wave

# Define the Bloch wavefunction
def bloch_wavefunction(x):
    return u_k(x)

# Define the spatial grid
x = np.linspace(0, L, 500)

# Calculate the Bloch wavefunction values
psi_k = bloch_wavefunction(x)

# Plot the real and imaginary parts of the Bloch wavefunction
plt.figure()
plt.plot(x, np.real(psi_k), label='Real')
plt.plot(x, np.imag(psi_k), label='Imaginary')
plt.xlabel('Position (x)')
plt.ylabel('Wavefunction Amplitude')
plt.legend()
plt.title(f'Bloch Wavefunction with k = {k:.2f}')
plt.grid(True)
plt.show()
```

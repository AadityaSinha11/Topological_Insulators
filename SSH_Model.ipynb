```python
 import numpy as np
import matplotlib.pyplot as plt
import cmath

#hopping potential
t1 = 2
t2 = 1

#Cell spacing
a = 1

#onsite energy
e1 = 0

#number of lattice sites
N = 4


#total length of chain


k1=np.linspace(-np.pi, np.pi,90)
ex1 = e1 - np.sqrt((t1)**2 + (t2)**2 + 2.*(t1)*(t2)*np.cos(k1*a))
ex2 = e1 + np.sqrt((t1)**2 + (t2)**2 + 2.*(t1)*(t2)*np.cos(k1*a))
```

```python
xticks = np.linspace(-np.pi/a, np.pi/a,90)
xlabels = ['' for k in xticks]
xlabels[0] = '$-\\pi$'
xlabels[-1] = '$\\pi$'
fig, axs = plt.subplots()
axs.set_xlim(-np.pi,np.pi)
axs.set_title('SSH Model')
axs.set_ylabel('$E$')
axs.set_xlabel('$ka$')
axs.set_yticks([e1-2.*t2, e1-t2, e1, e1+t2, e1+2.*t2])
axs.set_yticklabels(['$E_0-2t$', '$E_0-t$', '$E_0$', '$E_0+t$', '$E_0+2t$'])
axs.set_xticks(xticks)
axs.set_xticklabels(xlabels)
axs.grid(True)
axs.plot(k1, ex1)
axs.plot(k1,ex2)
axs.legend()
```

```python


# Compute the parametric equations for winding number
x = t1 + t2 * np.cos(k1)
y = t2 * np.sin(k1)

# Create the plot
plt.figure(figsize=(6, 6))
plt.plot(x, y, label='Circle')
plt.scatter(t1, 0 , color='red', label='Center')
plt.axvline(x=0, c="black", label="x=0")
plt.axhline(y=0, c="black", label="y=0")
plt.xlabel('hx')
plt.ylabel('hy')
plt.title('Winding number')
plt.axis('equal')
plt.show()
```
